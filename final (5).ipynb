```python
import numpy as np
```

```python
class Header:
    # helper methods to print lines for better output look
    @staticmethod
    def one_line(title):
        print(f"\n{'-'*20} {title} {'-'*20}")
        
    @staticmethod
    def two_line(title):
        print(f"\n{'='*20} {title} {'='*20}")
```

```python
class Person:
    def __init__(self, name, gender):
        self.name = name    # store person name
        self.gender = gender # store person gender
        
    @property
    def name(self):    # getter for name
        return self.__name
        
    @name.setter
    def name(self, name):   # setter for name with a check
        if isinstance(name, str) and len(name) > 0: # name cannot be empty
            self.__name = name
        else:
            raise ValueError("Name must be a non-empty string")
            
    @property
    def gender(self):   # getter for gender
        return self.__gender
        
    @gender.setter
    def gender(self, gender):   # setter for gender with a check
        if gender in ['male', 'female']: # must be male or female
            self.__gender = gender
        else:
            raise ValueError("Gender must be 'male' or 'female'")
            
    def info(self):
        return f"Person: {self.name} ({self.gender})"
```

```python
class Course:
    def __init__(self, course_name, course_unit):
        self.__course_name = course_name # private variable for name
        self.__course_unit = course_unit # private variable for unit

    @property
    def course_name(self):  # get course name
        return self.__course_name
        
    @property
    def course_unit(self):      # get course unit number
        return self.__course_unit

    def display_info(self):
        return f'Course: {self.__course_name} | Unit: {self.__course_unit}'
```

```python
class Student(Person):
    def __init__(self, name, gender, student_id):
        super().__init__(name, gender) # call the parent class constructor
        self.student_id = student_id # set id attribute
        self.__grades = {} # dictionary to save grades
        self.__units = 0 # counter for total units

    @property
    def student_id(self): # get student id
        return self.__student_id
        
    @student_id.setter
    def student_id(self, student_id): # set student id
        if isinstance(student_id, int) and student_id > 0: # check if it is positive integer
            self.__student_id = student_id
        else:
            raise ValueError("student_id must be integer")
            
    @property
    def units(self): # get total units
        return self.__units

    @property
    def grades(self): # get grades dictionary
        return self.__grades

    def add_grade(self, course_name, grade, course_unit):
        if isinstance(grade, (int, float)) and 0 <= grade <= 20: # grade must be between 0 and 20
            if course_name not in self.__grades.keys():
                self.__units += course_unit # only add unit if course is new
            self.__grades[course_name] = grade
        else:
            raise ValueError("Grade must be a number between 0 and 20")

    def get_average(self):
        if len(self.__grades) == 0:
            return 0
        return sum(self.__grades.values()) / len(self.__grades) # calculate simple average

    def show_transcript(self):
        Header.one_line(f'Transcripts for {self.name} (ID:{self.student_id})')
        print(f'Units: {self.__units}')
        if len(self.__grades) == 0:
            print("no grades recorded yet")
        else:
            for course, grade in self.__grades.items():
                print(f'Course: {course:<20} | Grade: {grade}')
            print(f'Average: {self.get_average():.2f}')
            
    def __str__(self):
        return f'Student: {self.name} | ID: {self.student_id} | Gender: {self.gender} | Units: {self.units} | Average grade: {self.get_average():.2f}'
```

```python
class EducationalSystem:
    def __init__(self):
        self.__students = [] # list for storing students
        self.__courses = [] # list for storing courses
        
    @staticmethod
    def sort_by_name(students): 
        # we use selection sort here to sort by name
        L = students.copy()
        for i in range(len(L)-1):
            for j in range(i+1, len(L)) :
                if L[j].name < L[i].name:
                    L[i], L[j] = L[j], L[i]
        return L

    @staticmethod
    def sort_by_id(students):   
        # bubble sort to sort students by their id
        arr = students.copy()
        n = len(arr)
        for i in range(n):
            for j in range(0, n-i-1):
                if arr[j].student_id > arr[j+1].student_id:
                    arr[j], arr[j+1] = arr[j+1], arr[j]
        return arr

    @staticmethod
    def sort_by_grade(students):   
        # bubble sort to sort students by their average grade
        arr = students.copy()
        n = len(arr)
        for i in range(n):
            for j in range(0, n-i-1):
                if arr[j].get_average() > arr[j+1].get_average():
                    arr[j], arr[j+1] = arr[j+1], arr[j]
        return arr

    @staticmethod
    def sort_by_unit(students):   
        # bubble sort to sort students by total units
        arr = students.copy()
        n = len(arr)
        for i in range(n):
            for j in range(0, n-i-1):
                if arr[j].units > arr[j+1].units:
                    arr[j], arr[j+1] = arr[j+1], arr[j]
        return arr
    
    def show_sorted_students(self, sort_type='name'):   # method to show sorted results
        if sort_type == 'name':
            sorted_list = self.sort_by_name(self.__students)
        elif sort_type == 'id':
            sorted_list = self.sort_by_id(self.__students)
        elif sort_type == 'grade':
            sorted_list = self.sort_by_grade(self.__students)
        elif sort_type == 'unit':
            sorted_list = self.sort_by_unit(self.__students)
        else:
            print("Invalid sort type.")
            return
        Header.two_line(f"Students Sorted by {sort_type}")
        for s in sorted_list:
            print(s)

    def add_student(self, student):
        self.__students.append(student)
        print(f"Student '{student.name}' added successfully.")
        
    def remove_student(self, student_id):
        for s in self.__students:
            if s.student_id == student_id:
                self.__students.remove(s)
                print(f"Student < Name: {s.name} - ID: {student_id} > removed successfully.")
                return
        print(f"Student {student_id} not found.")

    def add_course(self, course):   
        self.__courses.append(course)
        print(f"Course '{course.course_name}' added to catalog.")

    def record_grade(self, student_id, course_name, grade):
        course_exists = False
        course_unit = 0
        for c in self.__courses:
            if c.course_name == course_name:
                course_exists = True
                course_unit = c.course_unit
                break
        if not course_exists:
            raise ValueError(f"Error: Course '{course_name}' is not defined in the system catalog!")
        
        for s in self.__students:
            if s.student_id == student_id:
                s.add_grade(course_name, grade, course_unit)
                return
        print(f"Student {student_id} not found.")

    def show_student_transcript(self, student_id):  
        for s in self.__students:
            if s.student_id == student_id:
                s.show_transcript()   
                return
        print(f"Student {student_id} not found.")

    def analyze_grades(self):
        # collect all student grades inside a list
        all_grades = []
        for s in self.__students:
            for grade in s.grades.values():
                    all_grades.append(grade)
        if not all_grades:
                print("No grades available in the system for analysis.")
                return
        # use numpy array to calculate things easily
        grades_array = np.array(all_grades)
        mean_grade = np.mean(grades_array)
        max_grade = np.max(grades_array)     
        min_grade = np.min(grades_array)       
        std_grade = np.std(grades_array)     
        passed_count = np.sum(grades_array >= 10)
        failed_count = np.sum(grades_array < 10)
        Header.two_line("NumPy Statistical Analysis")
        print(f"Class Average:        {mean_grade:.2f}")
        print(f"Highest Grade:        {max_grade:.2f}")
        print(f"Lowest Grade:         {min_grade:.2f}")
        print(f"Standard Deviation:   {std_grade:.2f}")
        print(f"Passed Grades (>=10):  {passed_count}")
        print(f"Failed Grades (<10):   {failed_count}")
```

```python
system = EducationalSystem()

course1 = Course("OOP", 3)
course2 = Course("Data Structures", 4)
course3 = Course("Database", 3)

system.add_course(course1)
system.add_course(course2)
system.add_course(course3)

student1 = Student("Hamed", "male", 105)
student2 = Student("Sina", "male", 102)
student3 = Student("Ana", "female", 103)
student4 = Student("Amir", "male", 106)

system.add_student(student1)
system.add_student(student2)
system.add_student(student3)
system.remove_student(106)

system.record_grade(105, "OOP", 18.5)
system.record_grade(105, "Data Structures", 9.0)
system.record_grade(102, "OOP", 14.0)
system.record_grade(103, "OOP", 19.5)
system.record_grade(103, "Data Structures", 20.0)
system.record_grade(103, "Database", 17.5)

system.show_student_transcript(105)
system.show_student_transcript(103)
system.show_student_transcript(101)
system.show_sorted_students('id')
system.analyze_grades()
```

# Final Project Report - Student Management System

### 1. Class Structure Description
The project is built using Object-Oriented Programming (OOP) concepts:
* **Person (Base Class):** This is the main base class that has basic details like name and gender.
* **Student (Derived Class):** This class inherits from Person. It adds some more details like student_id and grades dictionary. It also uses encapsulation with @property to check inputs.
* **Course:** A separate class to keep the name and units of each course.
* **EducationalSystem:** The main manager class. It handles lists of students and courses, adds them, and runs the sorting or analysis methods.
* **Header:** A simple helper class with static methods just to print clean titles and separator lines.

### 2. Sorting Algorithms Used
The system can sort the students list in different ways using these algorithms:
* **Selection Sort:** Used in `sort_by_name` to sort the students alphabetically. It looks for the smallest name and puts it at the beginning.
* **Bubble Sort:** Used in `sort_by_id`, `sort_by_grade`, and `sort_by_unit`. It compares adjacent items in the list and swaps them if they are out of order.

### 3. Time Complexity Analysis
* **Time Complexity:** Both Selection Sort and Bubble Sort use nested loops (a loop inside another loop). Because of this, the time complexity is $O(N^2)$ for both algorithms. This is okay for a small class size, but it can get slow if we have thousands of students.
* **Space Complexity:** The sorting methods use `.copy()` to make a duplicate of the students list before sorting. This keeps the original list safe, but it takes a little bit of extra memory space.

### 4. Statistical Analysis Performed
Instead of writing normal python loops to do math, the project uses the **NumPy** library to analyze the grades data. We collect all student grades into a numpy array and find these results:
* Class Average (using `np.mean`)
* Highest and Lowest grades (using `np.max` and `np.min`)
* Standard Deviation (using `np.std` to see the spread of marks)
* Total passed and failed students count (using conditional array checking)
